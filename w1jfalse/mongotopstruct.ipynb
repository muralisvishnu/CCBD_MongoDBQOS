```python
from pandas import read_csv
df=read_csv("mt9.csv")
df
```

```python
#df

  
options = ["ycsb.usertable","local.system.replset","local.oplog.rs","config.system.sessions","local.replset.election"] 
  
# selecting rows based on condition 
resdf = df[df['ns'].isin(options)] 

```

```python
resdf
#removing millisecond unit 
data= resdf.replace(to_replace ='ms', value = '', regex = True) 
data["total"]= data["total"].astype(float) 
data["read"]= data["read"].astype(float) 
data["write"]= data["write"].astype(float) 
```

```python
resdf=data
data.index= range(len(data))
list(data.iloc[1])
yread=[]
ywrite=[]
ytot=[]

srread=[]
srwrite=[]
srtot=[]

elread=[]
elwrite=[]
eltot=[]

opread=[]
opwrite=[]
optot=[]

ssread=[]
sswrite=[]
sstot=[]
d={}
for i in range(len(data)):
    row=list(data.iloc[i])
    if row[0]=="ycsb.usertable":
        yread.append(row[2])
        ywrite.append(row[3])
        ytot.append(row[1])
    elif row[0]=="config.system.sessions":
        ssread.append(row[2])
        sswrite.append(row[3])
        sstot.append(row[1])
    elif row[0]=="local.replset.election":
        elread.append(row[2])
        elwrite.append(row[3])
        eltot.append(row[1])
    elif row[0]=="local.oplog.rs":
        opread.append(row[2])
        opwrite.append(row[3])
        optot.append(row[1])
    elif row[0]=="local.system.replset":
        srread.append(row[2])
        srwrite.append(row[3])
        srtot.append(row[1])
    if row[4] not in d:
        d[row[4]]=0

    
```

```python
import pandas as pd
res=pd.DataFrame()
res = res.assign(yread=yread, ywrite=ywrite,ytot=ytot ) 
res=res.assign(srread=srread, srwrite=srwrite,srtot=srtot)

res=res.assign(ssread=srread, sswrite=sswrite,sstot=sstot)

res=res.assign(elread=elread, elwrite=elwrite,eltot=eltot)

res=res.assign(opread=opread, opwrite=opwrite,optot=optot)
res=res.assign(time=list(d.keys()))
```
