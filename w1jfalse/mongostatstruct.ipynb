```python


#!!!!!!!!!!!!!!!!!  manually delete the time column !!!!!!!!!!!


from pandas import read_csv
df=read_csv("C:\\Users\\Abishek\\Documents\\SEM_5\\4_BD\Project\\fianl\\qfiles_5mil_3mil_w1_jt\\m8.csv")

#removing *
df= df.replace(to_replace ='\*', value = '', regex = True) 
#emoving the % 
df = df.replace(to_replace ='\%', value = '', regex = True) 
  

df["insert"]= df["insert"].astype(float)

df["query"]= df["query"].astype(float)

df["update"]= df["update"].astype(float)

df["delete"]= df["delete"].astype(float)

df["getmore"]= df["getmore"].astype(float)

df["used"]= df["used"].astype(float)
df["dirty"]= df["dirty"].astype(float)
```

```python
arw=list(df["arw"])
qrw=list(df["qrw"])
cmd=list(df["command"])

arwa=[]
arwb=[]
qrwa=[]
qrwb=[]

cmda=[]
cmdb=[]
for i in range(len(arw)):
    l=list(map(int,arw[i].split("|")))
    arwa.append(l[0])
    arwb.append(l[1])
    
    l=list(map(int,qrw[i].split("|")))
    qrwa.append(l[0])
    qrwb.append(l[1])
    
    l=list(map(int,cmd[i].split("|")))
    cmda.append(l[0])
    cmdb.append(l[1])
    
```

```python
df=df.assign(cmda=cmda,cmdb=cmdb,arwa=arwa,arwb=arwb,qrwa=qrwa,qrwb=qrwb)
df=df.drop(["command","arw","qrw"],axis=1)
```

```python
df.to_csv('C:\\Users\\Abishek\\Documents\\SEM_5\\4_BD\\Project\\fianl\\python files\\m8new.csv') 
```
